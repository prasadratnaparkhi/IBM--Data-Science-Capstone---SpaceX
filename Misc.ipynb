```python
import matplotlib.pyplot as plt

# Data for launch sites and their corresponding success/failure rates (replace with actual data)
launch_sites = ['KSC LC-39A', 'CCAFS LC-40', 'VAFB SLC-4E', 'CCAFS SLC-40']
success_rates = [0.90, 0.85, 0.80, 0.88]  # Success rates as percentages
failure_rates = [1 - rate for rate in success_rates]  # Calculate failure rates

# Find the launch site with the highest success ratio
highest_success_index = success_rates.index(max(success_rates))
highest_success_site = launch_sites[highest_success_index]

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(success_rates, labels=launch_sites, autopct='%1.1f%%', startangle=140, explode=[0.1 if site == highest_success_site else 0 for site in launch_sites])
plt.title('Launch Site Success Ratio for SpaceX')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()

# Print the launch site with the highest success ratio
print(f"The launch site with the highest success ratio is {highest_success_site}.")

# Print failure rates
for site, failure_rate in zip(launch_sites, failure_rates):
    print(f"Failure rate for {site}: {failure_rate:.1%}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
# Function to create scatter plot
def create_scatter_plot(payload_range):
    selected_payloads = df[df['PayloadMass'].between(payload_range[0], payload_range[1])]
    plt.figure(figsize=(10, 6))
    for site in df['LaunchSite'].unique():
        site_data = selected_payloads[selected_payloads['LaunchSite'] == site]
        plt.scatter(site_data['PayloadMass'], site_data['Outcome'], label=site)
    plt.xlabel('Payload (kg)')
    plt.ylabel('Launch Outcome')
    plt.title('Payload vs Launch Outcome for SpaceX')
    plt.legend()
    plt.show()

# Range slider for selecting payload range
payload_range_slider = widgets.FloatRangeSlider(
    min=df['PayloadMass'].min(),
    max=df['PayloadMass'].max(),
    value=[df['PayloadMass'].min(), df['PayloadMass'].max()],
    description='Payload Range (kg):',
    continuous_update=False
)

# Interactive widget
widgets.interactive(create_scatter_plot, payload_range=payload_range_slider)
```
